```python
kmeans = KMeans(n_clusters=10, random_state=0)
clucters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8,3))
centers = kmeans.cluster_centers_.reshape(10,8,8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
from cProfile import label
import matplotlib.pyplot as plt
import numpy as np
X = np.array(
        [[7,8], [12,20], [17,19], [26,15] ,[32,37],[87,75],[73,85], [62,80], [73,60], [87,96],])
labels = range(1, 11)
plt.figure(figsize = (10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0], X[:,1], label='True Position')
for label, x, y in zip(labels, X[:,0], X[:,1]):
    plt.annotate(label, xy = (x,y), xytext=(-3,3), textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
from turtle import distance
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as pyplot
linked = linkage(X, 'single')
labelList = range(1, 11)
plt.figure(figsize=(10,7))
dendrogram(linked, orientation='top', labels=labelList, distance_sort='ascending', show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clucter = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
clucter.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c = clucter.labels_, cmap = 'rainbow')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# загрузим набор данных
url = r'https://raw.githubusercontent.com/lucko515/clustering-python/master/Customer%20in%20Mall%20clusterng/Mall_Customers.csv'
customer_data = pd.read_csv(url)
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(28,12), dpi=180)
plt.figure(figsize=(10,7))
plt.title('customer dendograms')
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
